```python
df = pd.read_csv('train.csv')

dff=df.drop(['LEAVE'], axis =1,)

dff.head()
```

```python
''' ONE HOT ENCODING '''
train = pd.get_dummies(dff, columns=['CONSIDERING_CHANGE_OF_PLAN'])
train = pd.get_dummies(train, columns=['REPORTED_SATISFACTION'])
train = pd.get_dummies(train, columns=['REPORTED_USAGE_LEVEL'])
train = pd.get_dummies(train, columns=['COLLEGE'])


train.head()
X = train
train.head()
```

```python
Y=df.LEAVE
print(Y.shape)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=0)
```
